## Modificar celdas dentro del dataframe

```python
mask_edad = df_users.age < 21
```

```python
df_users.loc[mask_edad,'edad_cat'] = 'jovenes'
```

```python
df_users
```

```python
mask_edad = df_users.age >= 21
```

```python
df_users.loc[mask_edad,'edad_cat'] = 'adultos'
```

```python
df_users
```

```python
pd.cut(x=df_users.age, bins=[0, 21, 100])
```

```python
pd.cut(x=df_users.age, bins=[0, 21, 100], labels=['jovenes', 'adultos'])
```

```python
mask_age = df_users.age > 18
```

```python
mask_gender = df_users.gender == 'M'
```

```python
df_users.loc[mask_age | mask_gender,[True, True, False, True, False]]
```

```python
mask_gender
```

```python
df_users.columns
```

```python
df_users.loc[mask_age | mask_gender, [True, True, False, True, False, False]]
```

```python
df_users.loc[mask_age | mask_gender,:]
```

```python
df_users.loc[mask_age & mask_gender,:]
```

```python
df_users['edad_categoria'] = pd.cut(df_users.age, bins=[0, 21, 65, 100], labels=['jovenes', 'adultos', 'ancianos'])
```

```python
df_users.edad_categoria.value_counts()
```
